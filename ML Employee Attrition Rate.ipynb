```python
import numpy as np
import pandas as pd
train_data=pd.read_csv("Train.csv")
```

```python
df=pd.DataFrame(train_data)
```

```python
#Total number of records
n_records=df.shape[0]
print("Total number of records = {}".format(n_records))
```

```python
at_rate=train_data['Attrition_rate']

#separating the target variable from the data
new_features=train_data.drop('Attrition_rate',axis=1)
```

```python
df.skew(axis=0,skipna=True)
```

```python
# Total NULL values present in each feature/column
df.isnull().sum()
```

```python
#Total NULL values present in each row.
df.isnull().sum(axis=1)
```

```python
'''
Age- 412
Time of service - 144
Work Life bal -11
VAR2 - 577
VAR4 - 656
'''

#Imputing the NULL values
new_features_imputed=new_features.copy()
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
custom=['Age','Time_of_service','Pay_Scale','Work_Life_balance','VAR2','VAR4']
new_features_imputed[custom] = my_imputer.fit_transform(new_features[custom])
new_features_imputed.head()
```

```python
df_new=pd.DataFrame(new_features_imputed)
```

df_new.isnull().sum()

```python
#Normalization
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
numerical=['Age','Time_of_service']
new_features_imputed[numerical]=scalar.fit_transform(new_features_imputed[numerical])
new_features_imputed.head()
```

```python
#Removing skewness only on selected features by applying logarithmic transformation
skewed=['Time_of_service']
new_features_logtrans=new_features_imputed.copy()
new_features_logtrans[skewed]=new_features_imputed[skewed].apply(lambda x: np.log(x+1))
```

```python
df_logtrans=pd.DataFrame(new_features_logtrans)
```

```python
df_logtrans.skew(axis=0,skipna=True)
```

```python
#HOT-ENCODING
features_final = pd.get_dummies(new_features_logtrans)

from sklearn.model_selection import train_test_split

# Split the 'features' and 'attrition' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_final, 
                                                    at_rate, 
                                                    test_size = 0.2, 
                                                    random_state = 0)
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
#Applying Linear Regression
from sklearn.linear_model import LinearRegression
#x=np.array(X_train).reshape((-1,1))
#y=np.array(y_train)
model_1=LinearRegression().fit(X_train,y_train)
```

```python
y_pred=model_1.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```
